```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = data[['Age']].values
y = data['EstimatedSalary'].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
poly_feat = PolynomialFeatures(degree=2)
X_poly = poly_feat.fit_transform(X)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
plt.scatter(X, y, color='red', label='Actual Data') 
plt.plot(X, lin_reg.predict(X), color='green', label='Linear Line') 
plt.plot(sorted(X), poly_reg.predict(poly_feat.fit_transform(sorted(X))), color='blue', label='Polynomial Curve') 
```

```python
plt.title('Age vs Salary (Beginner Model)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()


print("Agar age 40 ho toh prediction kya hogi?")
print("Linear Regression:", lin_reg.predict([[40]]))
print("Polynomial Regression:", poly_reg.predict(poly_feat.fit_transform([[40]])))
```
