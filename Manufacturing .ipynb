```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('manufacturing.csv')

```

```python
X = df[['Temperature (°C)']].values 
y = df['Quality Rating'].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
poly_feat2 = PolynomialFeatures(degree=2)
X_poly2 = poly_feat2.fit_transform(X)
model2 = LinearRegression().fit(X_poly2, y)
```

```python
poly_feat4 = PolynomialFeatures(degree=4)
X_poly4 = poly_feat4.fit_transform(X)
model4 = LinearRegression().fit(X_poly4, y)
X_grid = np.arange(min(X), max(X), 0.1).reshape(-1, 1)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='red', alpha=0.3, label='Actual Data') # Manufacturing data points
```

```python
plt.plot(X, lin_reg.predict(X), color='green', label='Linear Regression')
```

```python
plt.plot(X_grid, model2.predict(poly_feat2.fit_transform(X_grid)), color='blue', label='Degree 2 (Polynomial)')
```

```python
plt.plot(X_grid, model4.predict(poly_feat4.fit_transform(X_grid)), color='black', label='Degree 4 (Polynomial)')
```

```python
plt.title('Manufacturing: Temperature vs Quality Rating')
plt.xlabel('Temperature (°C)')
plt.ylabel('Quality Rating')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sample_temp = [[200]]
print(f"Agar Temperature 200°C ho toh Quality Rating:")
print(f"Linear Prediction: {lin_reg.predict(sample_temp)[0]:.2f}")
print(f"Polynomial (D4) Prediction: {model4.predict(poly_feat4.fit_transform(sample_temp))[0]:.2f}")
```
